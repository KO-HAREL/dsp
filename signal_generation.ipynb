```python
import thinkdsp
import matplotlib.pyplot as plt

# Generate a sine wave signal
sine_wave = thinkdsp.SinSignal(freq=5)
wave = sine_wave.make_wave(duration=1, framerate=500)

# Plot the sine wave
wave.plot()
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import thinkdsp

# Set up subplots for all signals
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# Unit Sample Sequence (Discrete Delta)
n = np.arange(0, 10)
unit_sample = np.where(n == 0, 1, 0)
axs[0, 0].stem(n, unit_sample, use_line_collection=True)
axs[0, 0].set_title('Unit Sample Sequence')
axs[0, 0].set_xlabel('n')
axs[0, 0].set_ylabel('Amplitude')

# Unit Step Signal
unit_step = np.where(n >= 0, 1, 0)
axs[0, 1].stem(n, unit_step, use_line_collection=True)
axs[0, 1].set_title('Unit Step Signal')
axs[0, 1].set_xlabel('n')
axs[0, 1].set_ylabel('Amplitude')

# Unit Ramp Signal
unit_ramp = np.where(n >= 0, n, 0)
axs[1, 0].stem(n, unit_ramp, use_line_collection=True)
axs[1, 0].set_title('Unit Ramp Signal')
axs[1, 0].set_xlabel('n')
axs[1, 0].set_ylabel('Amplitude')

# Exponential Signal
t = np.linspace(0, 1, 500)
A = 1  # Amplitude
alpha = 5  # Exponential decay constant
exponential_signal = A * np.exp(-alpha * t)
axs[1, 1].plot(t, exponential_signal)
axs[1, 1].set_title('Exponential Signal')
axs[1, 1].set_xlabel('Time (s)')
axs[1, 1].set_ylabel('Amplitude')

# Square Wave Signal
square_wave = thinkdsp.SquareSignal(freq=5)
duration = 1
wave = square_wave.make_wave(duration=duration, framerate=500)
axs[2, 0].plot(wave.ts, wave.ys)
axs[2, 0].set_title('Square Wave Signal')
axs[2, 0].set_xlabel('Time (s)')
axs[2, 0].set_ylabel('Amplitude')

# Triangular Signal
triangular_wave = thinkdsp.TriangleSignal(freq=5)
wave = triangular_wave.make_wave(duration=duration, framerate=500)
axs[2, 1].plot(wave.ts, wave.ys)
axs[2, 1].set_title('Triangular Signal')
axs[2, 1].set_xlabel('Time (s)')
axs[2, 1].set_ylabel('Amplitude')

# Adjust layout
plt.tight_layout()
plt.show()
```
